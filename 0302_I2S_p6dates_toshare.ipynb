# Purpose of the script


This code reads activity data from an Excel file, cleans and structures the Work Breakdown Structure (WBS) into hierarchical levels, and then writes the updated data back into the same Excel sheet.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl

# General file path (update this to your local path)
file_path = r'C:\ProjectData\P6'
file_name = 'P6.xlsx'
sheet_name = 'Activities'

# Load Excel file
df = pd.read_excel(fr'{file_path}\{file_name}', sheet_name)

# Clean WBS column
df['WBS clean'] = df['WBS'].str.lstrip()

# Initialize WBS level columns
for i in range(1, 11):
    df[f'WBS level {i}'] = ""

# Dictionary to track latest WBS name at each level
wbs_levels = {}

# Populate WBS level columns based on Wbs Level
for idx, row in df.iterrows():
    level = int(row['Wbs Level'])
    wbs_levels[level] = row['WBS clean']
    for i in range(1, level + 1):
        df.at[idx, f'WBS level {i}'] = wbs_levels.get(i, "")

# Export back to Excel
full_path = fr'{file_path}\{file_name}'
wb = openpyxl.load_workbook(full_path)
ws = wb[sheet_name]

# Convert DataFrame to list of lists (including header)
data = [df.columns.tolist()] + df.values.tolist()

# Write data starting from row 1
for i, row in enumerate(data):
    for j, value in enumerate(row):
        ws.cell(row=i+1, column=j+1, value=value)

# Save workbook
wb.save(full_path)


```
