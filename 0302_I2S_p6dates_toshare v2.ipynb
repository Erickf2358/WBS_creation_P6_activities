# Purpose of the script


This code reads activity data from an Excel file, cleans and structures the Work Breakdown Structure (WBS) into hierarchical levels, and then writes the updated data back into the same Excel sheet.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl

# General file path (update this to your local path)
file_path = r'D:\2026\App testing'
file_name = 'P6.xlsx'
sheet_name = 'Activities'
# Load Excel file
df = pd.read_excel(fr'{file_path}\{file_name}', sheet_name)
```

```python
# Clean WBS column
prefix = 'I2S-7688-MP-74.'

df['WBS clean'] = (
    df['WBS']
    .str.lstrip()
    .str.replace(f'^{prefix}', '', regex=True)
)


# --- NEW: determine max level dynamically ---
max_level = int(df['Wbs level'].max())

# Initialize WBS level columns dynamically
for i in range(1, max_level):
    df[f'WBS level {i}'] = ""

# Dictionary to track latest WBS name at each level
wbs_levels = {}

#Populate WBS level columns
for idx, row in df.iterrows():
    level = int(row['Wbs level'])
    wbs_levels[level] = row['WBS clean']
    for i in range(1, level + 1):
        df.at[idx, f'WBS level {i}'] = wbs_levels.get(i, "")

# Export
output_file_name = 'P6_processed.xlsx'
output_path = fr'{file_path}\{output_file_name}'
df.to_excel(output_path, sheet_name=sheet_name, index=False)

```
